```python
from keras.preprocessing.text import one_hot
```

```python
sent = ['Well done!',
        'Good work',
        'Great effort',
        'nice work',
        'Excellent!',
        'Weak',
        'Poor effort!',
        'not good',
        'poor work',
        'Could have done better.']
```

```python
#vocablary_size
voc_size = 10000
```

## One Hot Representation

```python
onehot_Resp = [one_hot(word,voc_size) for word in sent]
print(onehot_Resp)
```

### Word Embedding Representation

```python
from keras.layers import Embedding
from keras.preprocessing.sequence import pad_sequences
from keras.models import Sequential
```

```python
sent_length= 5
embed_docs = pad_sequences(sequences=onehot_Resp , maxlen=sent_length , padding='pre')
print(embed_docs)
```

```python
# dimension 
dim= 10
```

```python
model = Sequential()
model.add(Embedding(voc_size , dim,input_length =  sent_length))
model.compile('adam','mse')
```

```python
model.summary()
```

```python
print(model.predict(embed_docs))
```
